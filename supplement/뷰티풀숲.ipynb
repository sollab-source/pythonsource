```python
import requests
from bs4 import BeautifulSoup

url = "https://finance.naver.com/"

response = requests.get(url)
```

```python
response.status_code
```

```python
# HTML 코드 엄청 기니 실행에 주의
# response.text
```

```python
soup = BeautifulSoup(response.text, "html.parser")

# HTML 예쁘게 출력하기
soup.prettify
```

```python
news_div = soup.select_one(".section_strategy ul")
news_titles = news_div.select("a")

for title in news_titles:
    _, office_id, article_id = [
        param.split("=")[1] for param in title["href"].split("&")
    ]
    url = f"https://n.news.naver.com/mnews/article/{office_id}/{article_id}"
    print(f"title: {title.text}, url: {url}")
```

```python
news_div = soup.find("div", class_="section_strategy").find("ul")
news_titles = news_div.find_all("a")

for title in news_titles:
    _, office_id, article_id = [
        param.split("=")[1] for param in title["href"].split("&")
    ]
    url = f"https://n.news.naver.com/mnews/article/{office_id}/{article_id}"
    print(f"title: {title.text}, url: {url}")
```
