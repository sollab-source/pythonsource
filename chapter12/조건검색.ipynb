```python
import pandas as pd
```

```python
df = pd.read_excel('apt_20250224.xlsx', header = 12, usecols='B,F,G,J,L')
df.head()
```

```python
df.tail()
```

```python
# 열 전체 데이터 출력하기
df['거래금액(만원)']
```

```python
df = df.replace({',': ''}, regex=True)
df = df.apply(pd.to_numeric, errors='ignore')
df.head()
```

```python
# 조건별로 출력하기
df[df['거래금액(만원)'] > 100000].head()
```

```python
df['시군구'][df['거래금액(만원)'] > 100000]
```

```python
df[(df['거래금액(만원)'] < 100000) & (df['전용면적(㎡)'] < 130)]
```

```python
df['시군구'][(df['거래금액(만원)'] < 100000) | (df['전용면적(㎡)'] < 130)]
```

```python
#df.loc[원하는 행의 조건, 원하는 열의 조건]
df.loc[:10, ['거래금액(만원)']]
```

```python
df.loc[:, ['시군구','거래금액(만원)']][df['거래금액(만원)'] > 300000]
```

```python
#새로운 값 추가
df['면적대비금액'] = df['거래금액(만원)'] / df['전용면적(㎡)']
```

```python
df.loc[:10, ('거래금액(만원)','전용면적(㎡)','면적대비금액' )]
```

```python
#데이터 정렬
df.sort_values(by='면적대비금액').loc[:, ('면적대비금액','시군구')]
```

```python
#원본 데이터는 변경되지 않음
df.head()
```

```python
df.sort_values(by='면적대비금액', ascending=False).loc[:, ('면적대비금액','시군구')]
```

```python
df[df['거래금액(만원)']> 100000].sort_values(by='거래금액(만원)').loc[:, ('거래금액(만원)','면적대비금액','시군구','단지명', '전용면적(㎡)')]
```
