#### Text 설정
##### 1. 이름 설정 - xlabel(), ylabel()

```python
x = [1,2,3]
y = [2,4,8]
plt.plot(x,y)
plt.title("꺽은선 그래프")
plt.xlabel('X 축')
plt.ylabel('Y 축')
plt.show()
```

```python
plt.plot(x,y,marker='o')  

for idx, txt in enumerate(y):
    #(x좌표,y좌표,보여줄문자,ha(수평정렬),color)
    plt.text(x[idx],y[idx]+0.2,txt, ha='center', color='blue')  

# 차트 작성
plt.plot(1.25, 3, 'o')
# annotate : xy(화살표 시작 위치), xytext(annotate 문자열 위치), shrink(화살촉과 차트사이 간격)
plt.annotate('annotate', xy=(1.25, 3), xytext=(1.5, 5.5), arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

#### 기타 그래프
#### 1) bar chart

```python
fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
# 범례 레이블
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

plt.bar(fruits, counts, label=bar_labels, color=bar_colors)

plt.ylabel('fruit supply')
plt.title('Fruit supply by kind and color')
plt.legend(title='Fruit color')

plt.show()
```

#### 2) Stacked bar chart

```python
species = (
    "Adelie\n $\\mu=$3700.66g",
    "Chinstrap\n $\\mu=$3733.09g",
    "Gentoo\n $\\mu=5076.02g$",
)
weight_counts = {
    "Below": np.array([70, 31, 58]),
    "Above": np.array([82, 37, 66]),
}
width = 0.5

bottom = np.zeros(3)

for boolean, weight_count in weight_counts.items():
    p = plt.bar(species, weight_count, width, label=boolean, bottom=bottom)
    bottom += weight_count

plt.title("Number of penguins with above average body mass")
plt.legend(loc="upper right")

plt.show()
```

#### 3) Pie Chart

```python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]

plt.pie(sizes, labels=labels)
plt.show()
```

```python
import pandas as pd

# 그래프 작성을 위해 데이터 로드
student_df = pd.read_csv("./students_data.csv")
student_df
```

```python
# 한 개의 subplots 생성
fig, ax = plt.subplots()
ax.bar(student_df['성명'], student_df['국어'],color="g")
ax.set_title('학생별 국어 점수')
plt.show()
```

```python
# y축을 공유하는 subplots 생성(1행, 2열)

fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(12,4))
ax1.bar(student_df['성명'], student_df['수학'],color="g")
ax1.set_title('학생별 수학 점수')

ax2.bar(student_df['성명'], student_df['과학'])
ax2.set_title('학생별 과학 점수')
plt.show()
```

```python
fig, axes = plt.subplots(2,2,figsize=(15,10)) # 2행 2열
fig.suptitle('여러 개의 그래프') # 4개를 관리하는 전체 타이틀

# 첫번째 그래프 작성
axes[0,0].bar(student_df['성명'],student_df['국어'],label='국어점수')
axes[0,0].set_title('학생별 국어 점수')
axes[0,0].legend()
axes[0,0].set(xlabel='성명',ylabel='점수')
axes[0,0].set_facecolor('lightyellow') # 전경색
axes[0,0].grid(linestyle='--', linewidth=0.5)


# 두번째 그래프 작성
axes[0,1].plot(student_df['성명'],student_df['영어'],label='영어점수')
axes[0,1].plot(student_df['성명'],student_df['수학'],label='수학점수')
axes[0,1].set_title('학생별 영어/수학 점수')
axes[0,1].legend()
axes[0,1].set(xlabel='성명',ylabel='점수')


# 세번째 그래프 작성
axes[1,0].barh(student_df['성명'],student_df['과학'])
axes[1,0].set_title('학생별 과학점수')
axes[1,0].set(xlabel='키',ylabel='성명')

# 네번째 그래프 작성
axes[1,1].plot(student_df['성명'], student_df['체육'], color='green', alpha=0.5)
axes[1,1].set(xlabel='성명',ylabel='점수')
plt.show()
```
