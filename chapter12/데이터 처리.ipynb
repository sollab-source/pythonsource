```python
import pandas as pd
import numpy as np
```

```python
date = pd.date_range('2025-01-01', periods=5, freq='Y')

df = pd.DataFrame(
    {
        "A": pd.Series(np.random.randint(0, 10, size=5), index=date),
        "B": pd.Series(np.random.randint(0, 10, size=5), index=date),
        "C": pd.Series(np.random.randint(0, 10, size=5), index=date),
        "D": pd.Series(np.random.randint(0, 10, size=5), index=date),
    }
)

df
```

```python
#행과 열 추가하기
df["E"] = pd.Series(np.random.randint(0,10, size=5), index=date)
df
```

```python

df.loc['2030-12-31', :] = [0,0,0,0,0]
df
```

```python
#임의 위치에 열 추가하기
df.insert(2, 'AA', 
  pd.Series(
    np.random.randn(10), 
    index=pd.date_range('1/1/2025', periods=10, freq='Y')
  )
)

df
```

```python
#행/열 삭제하기
df = df.drop('2025-12-31', axis=0)  #반환값을 다시 받아주어야함
df = df.drop('E', axis=1)
df
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'], 
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                   index=[4, 5, 6, 7])
```

```python
result = pd.concat([df1, df2])
result
```

```python
result = pd.concat([df1, df2], keys=['x', 'y'])
result
```

```python
result.index
```

```python
df3 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'], 
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                   index=[2, 3, 6, 7])

result = pd.concat([df1, df3], axis=1)
result
```

```python
result = pd.concat([df1, df3], axis=1, join='inner') 
result
```

```python
left = pd.DataFrame({'key': ['K0', 'K4', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
pd.merge(left, right, on='key')
```

```python
pd.merge(left, right, how='left', on='key')
```

```python
pd.merge(left, right, how='right', on='key')
```

```python
pd.merge(left, right, how='outer', on='key')
```

```python
pd.merge(left, right, how='inner', on='key')
```
