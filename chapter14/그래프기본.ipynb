#### [실습1] 기본 그래프 작성

```python
# y축 좌표 설정 / x 축은 0 ~ N-1 로 그려짐 / 선 스타일,마커,색상 기본 형식
plt.plot([-1,1,3,4])
plt.title("그래프")
plt.show()
```

#### [실습3] 딕셔너리를 이용한 그래프 작성

```python
data_dict = {'data_x': [1, 2, 3, 4, 5], 'data_y': [2, 3, 5, 10, 8]}
plt.plot('data_x', 'data_y', data=data_dict)
plt.show()
```

#### [실습5] 선, 마커, 마커 스타일, 색상 지정

```python
plt.plot([1, 2, 3], [1, 2, 3], marker="*", linewidth=2, linestyle="--", color="r")
plt.show()
```

```python
plt.plot([1, 2, 3], [1, 2, 3], "r*--", linewidth=2)
plt.show()
```

#### [실습6] 포맷 문자열을 이용한 스타일 지정

### 라벨, 범례, 주석

#### [실습] 라벨, 범례

```python
#  라벨, 범례

plt.plot([1,2,3,4,5], color = 'blue', label = 'data1')
plt.legend()
plt.show()
```

```python
#  라벨, 범례 - legend() 의 labels 속성 이용

plt.plot([1,2,3,4,5], color = 'blue')
plt.legend(labels= ['data1'])
plt.show()
```

```python
x = [1,2,3]
y = [2,4,8]
plt.plot(x, y, label="데이터")
plt.legend(loc="lower right")
plt.show()
```

```python
# Figure 객체 
fig = plt.figure(figsize=(10,5)) 
# super title : 여러개의 그래프를 그릴 때 전체 제목 지정
fig.suptitle("전체 그래프 제목")

# 첫번째 그래프 작성
ax1 = fig.add_subplot(1,2,1) # 1행 2열, 첫번째 컬럼에 차트 위치
ax1.plot(x,y)
ax1.set_title("첫번째")


x1 = [4,5,6]
y1 = [10,12,15]
# # 각 그래프 별 제목 설정
ax2 = fig.add_subplot(1,2,2)  # 1행 2열, 두번째 컬럼에 차트 위치
ax2.plot(x,y,c="r")
ax2.set_title("두번째")
plt.show()
```
